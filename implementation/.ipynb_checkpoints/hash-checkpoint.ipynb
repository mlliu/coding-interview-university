```python
#leetcode question
#Given string s and t which only consists of lowercase letter
#string t is generated by random shuffing s, and the add one more letter at random position
#find the letter that was added in t
```

```python
#solution
#build hashtable for s
#每个英文字母的ascii编码作为key,每个字母出现的次数作为value 
#m为26,作为prehash,不做进一步的序列压缩
#暴力搜索　复杂度是O(n2),使用hash复杂度将为O(n)
```

```python
from __future__ import print_function #解决print问题
class myerror(Exception):
    pass

class hash(object):
    #simple hash , use array,m is the size of hash table
    def __init__(self,m):
        self.m = m
        self.hashtable = [0]*self.m
    
    #define hash function use division method
    #hashvalue就是在hashtable的位置
    def hash_funcation(self,key):
        hashvalue = key % self.m #division method
        return hashvalue
    
    #如果出现一次就加１
    def insert(self,key):
        
        self.hashtable[self.hash_funcation(key)] += 1
        
    def delete(self,key):
        if self.hashtable[self.hash_funcation(key)] == 0:
            raise myerror("the item is not exist, can no do delete operation")
        self.hashtable[self.hash_funcation(key)] -= 1
        
    def search(self,key):
        return self.hashtable[self.hash_funcation(key)]
    
```

```python
if __name__ == "__main__":
    #先简单版，没有letter的重复出现
    s = "abcd"
    t = "aabcd"
    
    #build hashtable for s
    hashtable_size = 26#prehash，根据有２６个英文字母
    myhash = hash(hashtable_size)
    for char in s:
        myhash.insert(ord(char))
    
    output = None
    #search
    for char in t:
        #如果个数为０，说明这个就是ｓ里面不存在的
        if myhash.search(ord(char)) == 0:
            #output.join(char)
            output = char
           
            
        #如果不为０，说明出现一次，就把value减少一次
        else:
            myhash.delete(ord(char))
            
    print(output)
    
```
